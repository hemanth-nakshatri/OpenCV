```python
import cv2
import numpy as np

orange_pen = [3, 145, 115, 17, 255, 255]
green_pen = [31, 103, 58, 75, 255, 255]

orange_color = [0, 165, 255]
green_color = [0,150,0]

myColors = [orange_pen, green_pen]
myColorValues = [orange_color, green_color]

myPoints = [] ##[x, y, colorId]

video = cv2.VideoCapture(0)


def findColor(img, myColors, myColorValues):
    imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    count = 0
    newPoints = []
    for color in myColors:
        lower = np.array(color[0:3])
        upper = np.array(color[3:6])
        mask = cv2.inRange(imgHSV, lower, upper)
#         cv2.imshow(str(color[0]), mask)
        x, y = getContours(mask)
        cv2.circle(imgResult, (x,y), 15, myColorValues[count], cv2.FILLED)
        if x!=0 and y!=0:
            newPoints.append([x,y,count])
        count += 1
    return newPoints

def getContours(img):
    x,y,w,h = 0,0,0,0
    contours, hierarchy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for cnt in contours:
        # get Area of contour
        area = cv2.contourArea(cnt)
        
        if area > (0.0004 * img.size):
#             cv2.drawContours(imgResult, cnt, -1, (255,0,0), cv2.FILLED)
            perimeter = cv2.arcLength(cnt, closed = True)
            approx = cv2.approxPolyDP(cnt, 0.01*perimeter, closed = True)
            x, y, w, h = cv2.boundingRect(approx)
    return x+w//2, y 


def drawOnImage(myPoints, myColorValues):
    for point in myPoints:
        cv2.circle(imgResult, (point[0],point[1]), 15, myColorValues[point[2]], cv2.FILLED)


while True:
    _, image = video.read()
    
    imgResult = image.copy()
    
    newPoints = findColor(image, myColors, myColorValues)
    if len(newPoints)!= 0:
        for newP in newPoints:
            myPoints.append(newP)
            
    if len(myPoints)!=0:
        drawOnImage(myPoints, myColorValues)
    
    cv2.imshow("video", imgResult)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cv2.destroyAllWindows()
        video.release()
        break
```
